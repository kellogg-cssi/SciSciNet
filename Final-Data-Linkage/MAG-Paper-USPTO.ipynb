```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
df_MAG_Sel_wRC = pd.read_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/df_MAG_Sel_wRC.csv", sep='\t')
df_MAG_Sel_wRC
# Sel_MAG_Ref_Cite.ipynb
```

```python
df_MAG_Patents = pd.read_csv('../../DataLake/Reliance_Science_Patenting/_pcs_mag_doi_pmid.tsv', sep='\t')
df_MAG_Patents
# Read papers with patents link records.
# Version v34.
# https://zenodo.org/record/5803985#.Yfbht_XMJhE
```

```python
len(df_MAG_Patents['magid'].unique())
```

```python
len(df_MAG_Patents['patent'].unique())
```

```python
df_MAG_Patents.rename(columns={'magid':'MAGID'},inplace=True)
df_MAG_Patents
```

```python
df_MAG_Sub2Primary = pd.read_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/df_MAG_Sub2Primary.csv", sep='\t')
df_MAG_Sub2Primary
# Read_MAG_Paper_Raw.ipynb
```

```python
df_MAG_Sub2Primary.rename(columns={'0':'MAGID','1':'New_MAGID'},inplace=True)
df_MAG_Sub2Primary
```

```python
%%time
df_MAG_Patents_Sub = df_MAG_Patents.loc[df_MAG_Patents['MAGID'].isin(df_MAG_Sub2Primary['MAGID']),:]
df_MAG_Patents_Sub
```

```python
df_MAG_Patents_Sub_Merge = df_MAG_Patents_Sub.merge(df_MAG_Sub2Primary, how='inner', on='MAGID')
df_MAG_Patents_Sub_Merge
```

```python
df_MAG_Patents_Sub_Merge.drop(columns=['MAGID'], inplace=True)
df_MAG_Patents_Sub_Merge
```

```python
df_MAG_Patents_Sub_Merge.rename(columns={'New_MAGID':'MAGID'}, inplace=True)
df_MAG_Patents_Sub_Merge
```

```python
%%time
df_MAG_Patents_Pri = df_MAG_Patents.loc[~df_MAG_Patents['MAGID'].isin(df_MAG_Sub2Primary['MAGID']),:]
df_MAG_Patents_Pri
```

```python
df_MAG_Patents_New = pd.concat([df_MAG_Patents_Pri, df_MAG_Patents_Sub_Merge], ignore_index=True)
df_MAG_Patents_New
```

```python
%%time
df_MAG_Patents_Final = df_MAG_Patents_New.drop_duplicates(subset=['MAGID','patent'])
df_MAG_Patents_Final
```

```python
len(df_MAG_Patents_Final['MAGID'].unique())
```

```python
%%time
df_MAG_Patents_Final.loc[df_MAG_Patents_Final['MAGID'].isin(df_MAG_Sel_wRC['MAGID']), :]
```

```python
%%time
df_MAG_PT_Full = df_MAG_Sel_wRC[['MAGID']].merge(df_MAG_Patents_Final[['reftype','confscore','MAGID','patent','uspto','wherefound']], how='inner', on='MAGID')
df_MAG_PT_Full
# Merge the Primary MAG papers with patents.
```

# Read MAG to Patent Count

```python
df_MAG_PT_Sel
```

```python
%%time
df_MAG_PT_cnt = df_MAG_PT_Sel.groupby(['MAGID'],as_index=False)['patent'].agg(['count']).reset_index().sort_values(['count'], ascending=False)
df_MAG_PT_cnt
# Counting how many patents cite a paper
```

```python
df_MAG_PT_cnt.rename(columns={'count':'Patent_count'}, inplace=True)
df_MAG_PT_cnt
```

```python
df_MAG_PT_cnt.to_csv('./Data-Linkage/MAGID_Patentcount.csv', sep='\t', index=False)
```

# Read MAGID Patent Count

```python
%%time
MAGID_Patentcount = pd.read_csv('./Data-Linkage/MAGID_Patentcount.csv', sep='\t')
MAGID_Patentcount
```

```python
%%time
MAGID_Patentcount1 = MAGID_Patentcount.groupby(['Patent_count'])['MAGID'].agg(
    ['count']).reset_index().sort_values(['count'], ascending=False)
MAGID_Patentcount1
```

```python
MAGID_Patentcount1.head(20)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
colors = [i for i in cm.get_cmap('tab20').colors]
fig, ax1 = plt.subplots(figsize=(12, 8))
x = np.arange(20) + 1
ax1.bar(x, 
        MAGID_Patentcount1.head(20)['count'].values,
        color=colors[2])
ax1.set_xlabel('Patent citation counts', fontsize=16)
ax1.set_ylabel('MAG Paper counts', fontsize=16)
ax1.set_yscale('log')
ax1.tick_params(axis='y', labelsize=12)
ax1.tick_params(axis='x', labelsize=12)
# ax1.set_xscale('log')
```
