```python
%%time
import json
with open("../Final-Data/Data-Tables-MAG_" + date_str +"/MAG_Sub2Primary_Dict.json", 'r') as fp:
    MAG_Sub2Primary_Dict = json.load(fp)

MAG_Sub2Primary_Dict = {int(k):int(v) for k,v in MAG_Sub2Primary_Dict.items()}
print(len(MAG_Sub2Primary_Dict))
MAG_Sub2Primary_Dict[2257060365]
# Read_MAG_Paper_Raw.ipynb
```
