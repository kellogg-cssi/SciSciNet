```python
import pandas as pd
date_str = '2021-12-06'
```

```python
%%time
df_MAG_Sel_wRC = pd.read_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/df_MAG_Sel_wRC.csv", sep='\t')
df_MAG_Sel_wRC
# Sel_MAG_Ref_Cite.ipynb
```

```python
%%time
MAGID_NIHcount = pd.read_csv('./Data-Linkage/MAGID_NIHcount.csv', sep='\t')
MAGID_NIHcount
# MAG-Paper-NIH.ipynb
```

```python
%%time
MAGID_Patentcount = pd.read_csv('./Data-Linkage/MAGID_Patentcount.csv', sep='\t')
MAGID_Patentcount
# MAG-Paper-USPTO.ipynb
```

```python
%%time
df_MAG_Newsfeed_cnt = pd.read_csv('./Data-Linkage/MAGID_Newsfeedcount.csv', sep='\t')
df_MAG_Newsfeed_cnt
# MAG-Paper-Newsfeed.ipynb
```

```python
%%time
df_MAG_Twitter_cnt = pd.read_csv('./Data-Linkage/MAGID_Twittercount.csv', sep='\t')
df_MAG_Twitter_cnt
# MAG-Paper-Twitter.ipynb
```

```python
%%time
df_MAG_NCT_cnt = pd.read_csv('./Data-Linkage/MAGID_NCTcount_wPMID_AACT.csv', sep='\t')
df_MAG_NCT_cnt
# MAG-Paper-NCT-ACCT.ipynb
```

```python
%%time
MAGID_NIHcount.loc[MAGID_NIHcount['MAGID'].isin(df_MAG_Sel_wRC['MAGID']), :]
```

```python
%%time
NIH_Patent_test = MAGID_NIHcount.merge(MAGID_Patentcount, how='outer', on='MAGID')
NIH_Patent_test
```

```python
%%time
MAGID_NIHcount.loc[MAGID_NIHcount['MAGID'].isin(NIH_Patent_test['MAGID']), :]
```

```python
%%time
MAGID_Patentcount.loc[MAGID_Patentcount['MAGID'].isin(NIH_Patent_test['MAGID']), :]
```

```python
%%time
NIH_Patent = MAGID_NIHcount.merge(MAGID_Patentcount, how='outer', on='MAGID')
NIH_Patent_News = NIH_Patent.merge(df_MAG_Newsfeed_cnt, how='outer', on='MAGID')
NIH_Patent_News_Twit = NIH_Patent_News.merge(df_MAG_Twitter_cnt, how='outer', on='MAGID')
NIH_Patent_News_Twit_NCT = NIH_Patent_News_Twit.merge(df_MAG_NCT_cnt, how='outer', on='MAGID')
NIH_Patent_News_Twit_NCT
```

```python
%%time
SciNet_MAG_Papers = df_MAG_Sel_wRC.merge(NIH_Patent_News_Twit_NCT, how='inner', on='MAGID')
# inner join for the focal papers.
SciNet_MAG_Papers
```

```python
%%time
SciNet_Papers = df_MAG_Sel_wRC.merge(NIH_Patent_News_Twit_NCT, how='left', on='MAGID')
# left join for the primary papers.
SciNet_Papers
```

```python
%%time
SciNet_Papers.loc[pd.notnull(SciNet_Papers['NIH_count']),:]
```

```python
SciNet_Papers.to_csv("../Final-Data/Data-Tables-MAG_" + date_str + "/SciNet_Papers_V0-1.csv", sep='\t', index=False)
```
