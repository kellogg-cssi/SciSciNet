# I. Read Raw MAG data

```python
date_str = "2021-12-06"
```

```python
%%time
import csv
import pandas as pd

df_MAG = pd.read_csv('../../MAG_'+date_str+'/mag/Papers.txt', sep='\t', header=None, quoting=csv.QUOTE_NONE, 
                      usecols=[0,2,3,7,8,11,12,13,22,24])
# https://docs.microsoft.com/en-us/academic-services/graph/reference-data-schema#papers
# Read MAG data.
# 0:MAG_PaperId; 2:DOI; 3:DocType; 7:Year; 8:Date; 11:JournalId; 12:ConferenceSeriesId; 13:ConferenceInstanceId;
# 18:ReferenceCount; 19:CitationCount; 20:EstimatedCitation; 22:FamilyId; 24:DocSubTypes.

df_MAG
```

```python
%%time
df_MAG = df_MAG.loc[((pd.notnull(df_MAG[3])) | (pd.notnull(df_MAG[2]))), :]
# Delete papers that had neither DOI nor DocType.

df_MAG = df_MAG.loc[(df_MAG[3] != 'Patent'), :]
# Delete papers with DocType as 'Patent'.

df_MAG
```

```python
%%time
df_MAG = df_MAG.loc[pd.isnull(df_MAG[24]), :]
# Delete papers retracted.

df_MAG
```

```python
%%time
df_MAG_cnt = df_MAG[[0,7]].groupby([7], as_index=False).size()
df_MAG_cnt
```

```python
df_MAG_cnt.rename(columns={7: "year"}, inplace=True)
df_MAG_cnt
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
sty = 'default'
mpl.style.use(sty)
fig, ax1 = plt.subplots()
ax1.plot("year", "size", '-', data = df_MAG_cnt)
ax1.set_yscale('log')
# ax1.set_xscale('log')
ax1.set_xlabel('Year',fontsize = 12)
ax1.set_ylabel('# of publications in MAG',fontsize = 12)
plt.xlim(right=2020)
plt.xticks(np.arange(1800, 2021, 10),rotation=300)
spacing = 0.400
fig.subplots_adjust(bottom=spacing)
# plt.xlim([1800, 2021])
fig.savefig("MAG_Figures/Raw/MAG_PubCount_Raw.pdf", bbox_inches='tight')
```

## Family Paper Mapping

```python
df_MAG_Family = df_MAG.loc[pd.notnull(df_MAG[22]) & df_MAG[22].isin(df_MAG[0]), :]
df_MAG_Family
```

```python
df_MAG_Family_Map = df_MAG_Family.loc[df_MAG_Family[22] != df_MAG_Family[0], :]
df_MAG_Family_Map
```

```python
df_MAG_Sub2Primary = pd.DataFrame(df_MAG_Family_Map[[0,22]].values)
df_MAG_Sub2Primary
```
