# Select MAG Primary Papers

```python
date_str = '2021-12-06'
```

```python
%%time
import pandas as pd
df_MAG_whole = pd.read_csv("./Data-Tables-MAG_" + date_str + "/df_MAG_whole.csv", sep='\t')
df_MAG_whole
```

```python
%%time
df_MAG_whole.loc[pd.notnull(df_MAG_whole['24']), :]
```

```python
%%time
df_MAG_whole.drop(['24'], axis=1, inplace=True)
df_MAG_whole
```

```python
%%time
df_MAG_Sel = df_MAG_whole.loc[((pd.isnull(df_MAG_whole['22'])) | (df_MAG_whole['22'] == df_MAG_whole['0'])), :]
df_MAG_Sel
```

## Save df_MAG_Sel as primary paper table.

```python
df_MAG_Sel.to_csv("./Data-Tables-MAG_" + date_str + "/df_MAG_Sel.csv", sep='\t', index=False)
```
