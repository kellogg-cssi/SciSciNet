```python
%%time
import csv
import pandas as pd

df_MAG_F = pd.read_csv('../../MAG_'+date_str+'/advanced/FieldsOfStudy.txt', sep='\t', header=None, 
                     quoting=csv.QUOTE_NONE)

df_MAG_F
```

```python
df_MAG_F_Lv01 = df_MAG_F.loc[df_MAG_F[5] < 2, :]
df_MAG_F_Lv01
```

```python
df_MAG_PF_Lv01 = df_MAG_PF.loc[df_MAG_PF[1].isin(df_MAG_F_Lv01[0]), :]
df_MAG_PF_Lv01
```

## Read Selected MAG Papers

```python
%%time
df_MAG_Sel = pd.read_csv("./Data-Tables-MAG_" + date_str + "/df_MAG_Sel.csv", sep='\t')
df_MAG_Sel
```

## Level 0

```python
df_MAG_F_Lv0 = df_MAG_F.loc[df_MAG_F[5] < 1, :]
df_MAG_F_Lv0
```

```python
df_MAG_PF_Lv0 = df_MAG_PF.loc[df_MAG_PF[1].isin(df_MAG_F_Lv0[0]), :]
df_MAG_PF_Lv0
```

```python
df_MAG_PF_Lv0_Sel = df_MAG_PF_Lv0.loc[df_MAG_PF_Lv0[0].isin(df_MAG_Sel['0']), :]
df_MAG_PF_Lv0_Sel
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()
plt.hist(df_MAG_PF_Lv0_Sel[2], log=True, bins=20)
plt.xlabel('Confidence score of Lv0-fields', fontsize=12)
plt.ylabel('# of papers', fontsize=12)
# fig.savefig("MAG_figures/Read_MAG_Paper_Fields-Lv0_field_confidence_score.pdf", bbox_inches='tight')
# Bigger number representing higher confidence.
```

```python
df_MAG_PF_Lv0_Sel
```

## Save df_MAG_PF_Lv0_Sel as Selected Papers-Fields-Lv0 with Conf Score.

```python
df_MAG_PF_Lv0_Sel.to_csv("./Data-Tables-MAG_" + date_str + "/df_MAG_PF_Lv0_Sel.csv", sep='\t', index=False)
```
