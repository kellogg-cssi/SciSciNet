```python
%%time
SciNet_Team = SciNet_PaperAuthorAffiliations.groupby(
    ['PaperID'],as_index=False)['AuthorID'].agg({'count': "nunique"})
SciNet_Team
```

```python
SciNet_Team.rename(columns={'count':'Team_Size'}, inplace=True)
SciNet_Team
```

# Institution size

```python
%%time
SciNet_PAA_Sel = SciNet_PaperAuthorAffiliations.loc[
    pd.notnull(SciNet_PaperAuthorAffiliations['AffiliationID']), :]
SciNet_PAA_Sel
```

```python
%%time
SciNet_Institution = SciNet_PAA_Sel.groupby(
    ['PaperID'],as_index=False)['AffiliationID'].agg({'count': "nunique"})
SciNet_Institution
```

```python
SciNet_Institution.rename(columns={'count':'Institution_Count'},inplace=True)
SciNet_Institution
```

```python
%%time
SciNet_Team_Institution = SciNet_Team.merge(SciNet_Institution, how='left', on='PaperID')
SciNet_Team_Institution
```
