```python
%%time
import csv
import pandas as pd

df_MAG_Affiliations = pd.read_csv('../../MAG_'+date_str+'/mag/Affiliations.txt', sep='\t', header=None, quoting=csv.QUOTE_NONE, 
                      usecols=[0,2,3,4,5,10,11,12])
df_MAG_Affiliations
```

```python
%%time
df_Affiliations_Sel = df_MAG_Affiliations.loc[
    df_MAG_Affiliations['AffiliationID'].isin(SciNet_PaperAuthorAffiliations['AffiliationID']), :]
df_Affiliations_Sel
```
