```python
%%time
SciNet_Papers = pd.read_csv('../Final-SciNet-Tables/SciNet_Papers-Temp0.tsv', sep='\t')
SciNet_Papers
```

```python
%%time
SciNet_Papers_DOI = SciNet_Papers.loc[pd.notnull(SciNet_Papers['DOI']),:]
SciNet_Papers_DOI
```

```python
%%time
SciNet_Papers_DOI = SciNet_Papers_DOI.copy()
SciNet_Papers_DOI['DOI'] = SciNet_Papers_DOI['DOI'].str.lower()
SciNet_Papers_DOI
```

```python
%%time
Dim_SciNet_Papers_DOI = Dim_Papers_DOI.merge(
    SciNet_Papers_DOI[['PaperID','DOI','Citation_Count','Reference_Count','Year']], 
    how='inner',left_on='doi',right_on='DOI')
Dim_SciNet_Papers_DOI
```

```python
%%time
import scipy.stats
C_Pearson = scipy.stats.pearsonr(Dim_SciNet_Papers_DOI['times_cited'].values, 
                                 Dim_SciNet_Papers_DOI['Citation_Count'].values)
C_Pearson
```

```python
C_Spearman = scipy.stats.spearmanr(Dim_SciNet_Papers_DOI['times_cited'].values, 
                                 Dim_SciNet_Papers_DOI['Citation_Count'].values)
C_Spearman
```

```python
%%time
R_Pearson = scipy.stats.pearsonr(Dim_SciNet_Papers_DOI['num_references'].values, 
                                 Dim_SciNet_Papers_DOI['Reference_Count'].values)
R_Pearson
```

```python
%%time
R_Spearman = scipy.stats.spearmanr(Dim_SciNet_Papers_DOI['num_references'].values, 
                                 Dim_SciNet_Papers_DOI['Reference_Count'].values)
R_Spearman
```
